git add - добавляет файлы в индекс

git commit -m "message"- сохранение изменений

git commit -a - добавление и коммит


git log  - история изменений

git checkout "commit hash" - переход на коммит в рабочей директории

git checkout master

git branch "branch name" - создание новой ветки

git checkout -b "branch name" - создание и переход в ветку (branch+checkout)

git branch - список всех веток

git branch -m - переименование текущей ветки

git branch -d - удаление



Если хотите создать новую ветку для сохранения созданных коммитов, можете
сделать это (сейчас или позже), используя команду switch с параметром -c.
Например:

  git switch -c <новая-ветка>
